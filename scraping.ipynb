```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd 
```

```python
url= 'https://awoiaf.westeros.org/index.php/List_of_characters#A'
headers = headers = {"Accept-Encoding":"gzip, deflate, br",
"Accept":"*/*",
"Connection":"close", "Upgrade-Insecure-Requests":"1",
"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
character_page = requests.get(url,headers=headers)
content = character_page.content
```

```python
soup = BeautifulSoup(content,"html.parser")
dom = etree.HTML(character_page.text)
```

```python
names_xpath = dom.xpath('//div[@class="mw-parser-output"]/ul/li/a[1]')
```

```python
names_xpath
```

```python
names= [n.text for n in names_xpath]
```

```python
print(names)
```

```python
pd.Series(names,name='charName').to_csv("characters.csv",index=False)
```

```python
#######houses###
url2 = 'https://awoiaf.westeros.org/index.php/Houses_of_Westeros'
housespage = requests.get(url2,headers=headers)
housespage_txt = housespage.text
dom_houses = etree.HTML(housespage_txt)
```

```python
houses_xpath = dom_houses.xpath('//table/tbody/tr/td/div/ul/li/a[1]')
```

```python
houses= [n.text for n in houses_xpath]
```

```python
pd.Series(houses,name='houseName').to_csv("houses.csv",index=False)
```

```python

df = pd.read_csv("characters.csv",use_nullable_dtypes=True)
```

```python
print(df.info())
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['charName'] = df['charName'].str.lower()
```

```python
# lady
def fname(row):
    if row.split(" ")[0] not in ["ser","prince","lady","lord","king","house"]:
        return row.split(" ")[0]
    else :
        if len(row.split(" "))>1 :
            return row.split(" ")[1]
        else:
            return []
df['firstName']= df['charName'].apply(fname)
df
```

```python
df[df["firstName"].duplicated()]
```

```python
df.drop(index=df[df["firstName"]=="the"].index,inplace=True)
```

```python
df.drop(index=df[df["firstName"].str.len()<2].index,inplace=True)
```

```python
#lady
```

```python
df.sort_values(by="firstName",inplace=True)
```

```python
df.info()
```

```python
df.to_csv("characters.csv",index=False)
```
